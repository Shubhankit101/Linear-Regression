```python
def hypo(x,theta):
    return theta[0]+theta[1]*x


def grad(x,y,theta):
    grad=np.zeros(2)
    m=x.shape[0]
    for i in range(m):
        hx=hypo(x[i],theta)
        grad[0]=(hx-y[i])
        grad[1]=(hx-y[i])*x[i]
    return grad


def gd(x,y,lr=0.1):
    theta=np.zeros(2)
    mx=300
    error_list=[]
    theta_list=[]
    for i in range(mx):
        gradient=grad(x,y,theta)
        theta[0]=theta[0]-lr*gradient[0]
        theta[1]=theta[1]-lr*gradient[1]
        e=error(x,y,theta)
        error_list.append(e)
        theta_list.append((theta[0],theta[1]))
    return theta,error_list,theta_list


def error(x,y,theta):
    m=x.shape[0]
    err=0
    for i in range(m):
        hx=hypo(x[i],theta)
        err+=(hx-y[i])**2
        
    return err/m


ft,error_list,theta_list=gd(x,y)


```
